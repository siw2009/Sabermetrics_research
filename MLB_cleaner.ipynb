```python
rlt = []
with open('./datasets/MLB_dataset.csv', 'r') as file:
    for line in file:
        data = line.strip().split(',')
        if len(data) < 2:  continue
        if data[1]:
            rlt.append(data[:17])
        else:
            rlt.append([data[0]])
            rlt[-1] += data[17:]
```

```python
rlt
```

```python
with open('./datasets/MLB_dataset_clean.csv', 'w') as file:  file.write('')
with open('./datasets/MLB_dataset_clean.csv', 'a') as file:
    for line in rlt:
        for x in line:
            file.write(f'{x},')
        file.write('\n')
```
